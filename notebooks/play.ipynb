```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use('_mpl-gallery')

data = pd.read_csv('datasets/responses.csv')
data.sort_values(by='Responded At')
data.count()
```

```python
id_col = 'Participant ID'
responded_col = 'Responded At'
platform_col = 'External Platform'

gender_col = '[Participant] Gender'
studies_col = '[Participant] Studies Level'
age_col = '[Participant] Age'
social_media_col = '[Participant] Social Media Usage'
reddit_col = '[Participant] Reddit Usage'
fake_news_col = '[Participant] Fake News Ability'
reddit_as_news_col = '[Participant] Reddit As News Source'
reddit_credibility_col = '[Participant] Reddit Credibility Evaluation'

credibility_col = '[Credibility] Credibility Rating'
upvote_col = '[Credibility] Would upvote'
upvote_why_col = '[Credibility] Would upvote reason'
topic_familiarity_col = '[Credibility] Topic familiarity'
content_style_col = '[Credibility] Content Style'
content_style_effect_col = '[Credibility] Content Style Effect'

post_upvotes_col = '[Post] Upvotes'
post_comments_col = '[Post] Comments'
post_title_col = '[Post] Title'
post_title_sent_col = '[Post] Title Sentiment'
post_link_domain_col = '[Post] Link Domain'
post_sub_col = '[Post] Subreddit'
post_awards_col = '[Post] Award Count'
post_ratio_col = '[Post] Upvote Ratio'
post_at_col = '[Post] Posted At'
```

```python
# Reddit usage by gender
data.groupby(by=[gender_col])[[reddit_col]].mean()
```

```python
# Reddit usage by age
reddit_by_age = data.groupby(by=[age_col])[reddit_col].mean().to_dict().items()
reddit_by_age = np.array(list(reddit_by_age))
x,y = reddit_by_age[:, 0], reddit_by_age[:, 1]

fig, ax = plt.subplots()
ax.bar(x, y)
```

```python
# Reddit usage by gender
data.groupby(by=[gender_col])[[reddit_col]].mean()
```

```python
# Ability to detect fake news by gender
data.groupby(by=[gender_col])[[fake_news_col]].mean()
```

```python
# Ability to detect fake news by age
data.groupby(by=[age_col])[[fake_news_col]].mean()
```
